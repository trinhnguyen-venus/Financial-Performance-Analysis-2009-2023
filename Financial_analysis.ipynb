## Import data

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import data
df = pd.read_csv("Financial Statements.csv")

df
```

## Data Understanding

```python
# Get basic info about dataset
print(df.info())
round(df.describe().T, 2)
```

```python
sns.boxplot(data = df["Market Cap(in B USD)"])
plt.title("Boxplot of Market Cap (in B USD)")
plt.show()
```

```python
# Fill missing value with median
df["Market Cap(in B USD)"].fillna(df["Market Cap(in B USD)"].median(), inplace = True)
```

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
# Standardize column names
# Rename all columns to lowercase and replace spaces with underscores
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.columns
```

## Exploratory Data Analysis (EDA)

```python
# Plot correlation between financial metrics
#plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap of Financial Variables")
plt.show()
```

```python
# Distribution of key financial metrics
numeric_cols = ["Market_Cap(in_B_USD)", "Revenue", "Gross_Profit", "Net_Income", "ROE", "ROA"]
df[numeric_cols].hist(bins=20, figsize=(10,6))
plt.suptitle("Distribution of Key Financial Metrics")
for ax in plt.gcf().axes:
    ax.grid(False)
plt.tight_layout()
plt.savefig("Images\Distribution_of_Key_Financial_Metrics.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12, 5))
# Correlation Market_cap vs Revenue
sns.scatterplot(data = df, x = "Revenue", y = "Market_Cap(in_B_USD)", ax = axes[0])
sns.regplot(data = df, x = "Revenue", y = "Market_Cap(in_B_USD)", ax = axes[0], scatter = True, color = "red")
axes[0].set_title("Market Cap vs Revenue")

# Correlation ROE vs Debt
sns.scatterplot(data = df, x = "Debt/Equity_Ratio", y = "ROE", ax = axes[1])
sns.regplot(data = df, x = "Debt/Equity_Ratio", y = "ROE", ax = axes[1], scatter = True, color = "red")
axes[1].set_title("ROE vs Debt-to-Equity Ratio")
plt.tight_layout()
plt.savefig("Images\Market_Cap_&_ROE_Relationship.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Market Cap vs Net Income by Category (Industry)
sns.scatterplot(data=df, x="Revenue", y="Market_Cap(in_B_USD)", hue = "Category")
plt.title("Relationship between Revenue and Market Capitalization")
plt.xlabel("Revenue (in Billion USD)")
plt.ylabel("Market Cap (in Billion USD)")
plt.savefig("Images\Revenue_vs_marketcap.png", dpi=300, bbox_inches='tight')
plt.show()
```

## FINANCIAL RATIOS ANALYSIS

```python
# Lineplot: Trend of ROE and ROA over time 
sns.lineplot(data = df, x = "Year", y = "ROE", label = "ROE")
sns.lineplot(data = df, x = "Year", y = "ROA", label = "ROA")
plt.title("Trend of ROE and ROA Over Time")
plt.xlabel("Year")
plt.ylabel("Percentage (%)")
plt.tight_layout()
plt.savefig("Images\Profitability_Trend.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
avg_metrics = df.groupby("Category")[["ROA", "ROE", "Debt/Equity_Ratio"]].mean().sort_values(ascending = False, by = ["ROA", "ROE"]).reset_index()
```

```python
# Barplot: Average financial indicators by category
avg_metrics_melted = avg_metrics.melt(id_vars="Category", var_name="Metric", value_name="Value")
sns.barplot(data = avg_metrics_melted, x = "Category", y = "Value", hue = "Metric")
plt.title("Average Financial Metrics by Industry")
plt.xlabel("Industry Category")
plt.ylabel("Average Value")
plt.xticks(rotation=45)
plt.show()
```

```python
# Create new ratio variables for deeper analysis
# Gross margin (profitability of core business):
df["Gross_Margin"] = df["Gross_Profit"] / df["Revenue"]

# Profit margin (overall profitability):
df["Profit_Margin"] = df["Net_Income"] / df["Revenue"]

# Debt to equity ratio (already available, but standardize name)
df["Debt_to_Equity"] = df["Debt/Equity_Ratio"]

# Return on capital (ROIC - includes debt and equity)
df["Return_on_Capital"] = df["Net_Income"] / (df["Share_Holder_Equity"] + df["Debt_to_Equity"])

# Revenue per employee (efficiency metric)
df["Revenue_per_Employee"] = df["Revenue"] / df["Number_of_Employees"]

df[["Company", "Revenue", "Gross_Profit", "Gross_Margin", "Profit_Margin", "Debt_to_Equity", "Return_on_Capital", "Revenue_per_Employee"]]
```

```python
# Summarize performance by catergory
industry_summary = df.groupby("Category")[["ROA", "ROE", "Debt_to_Equity", "Gross_Margin", "Profit_Margin",
                                          "Return_on_Capital", "Revenue_per_Employee"]].agg(["mean", "std"]).round(2)
industry_summary
```

```python
# --- Create a pivot table for quick comparison ---
pivot = df.pivot_table(values=["ROE","ROA","Debt_to_Equity","Profit_Margin"],index="Category", aggfunc="mean").round(2)
print(pivot)
```
